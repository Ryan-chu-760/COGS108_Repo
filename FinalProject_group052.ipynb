# 1. Introduction

### Overview

Every 4 years, we see countries cheerfully bidding higher and higher amounts in order to secure the spot as host nation for the FIFA World Cup, a beloved and well-known global event. World Cup host countries then spend millions of dollars, and years of effort in planning, organizing, and marketing to deliver a successful World Cup for players and audiences from all over the world. It is commonly assumed that by hosting the World Cup, countries can anticipate acquiring widespread prestige, influence, and socio-economic benefits, rendering their initial investment worthwhile.

__In this project, we focus on quantifying the real-world outcomes of countries that have hosted the World Cup by performing data analysis on 4 key factors that detail a country's relative prosperity: GDP, Unemployment, Tourism, and Crime. We analyze the change in these key factors over time in countries that have hosted the World Cup, and in countries that have not hosted the World Cup. We then compare our statistical results to assess if hosting the World Cup can be attributed to an increase in that country's overall prosperity (if their investments in time and effort proved to be beneficial).__

### Background
The World Cup is an internationally celebrated event followed by people, regardless of their home country and age, spurring about national unity and global excitement. Historically, the World Cup was only permitted to be hosted by developed nations. This changed in 2008 when China hosted the 2008 Summer Olympics. As a result, developing countries have had the opportunity to experience the potential benefits that come with hosting grand global events. Often times, it is perceived that being a host country is a favourable decision (especially for a developing country) as it opens the window for econonimic development, tourism, and increased business activity. However, there are various negative impacts of hosting an event like the World Cup such as greater government borrowing (leading to more debt), corruption, and inefficient allocation of the government budgets for the long term (Ludden 2018). 

Prior research has focused on predicting the costs and benefits of hosting a mega sports event. Some research has suggested that the overall benefit exceeds the costs while others have argued that the long term costs outweigh the benefits. Tourism associated with the 2010 World Cup has also been analyzed in order to determine the impact of World Cup associated tourism for a  developing country (Peeters, Matheson, and Szymansk 2014). Additionally, there have been studies examining the corruption within the Fédération Internationale de Football Association (FIFA) and the hosting governments (Ludden 2018). Due to the lack of consistency and agreement in whether or not hosting the World Cup is beneficial for a developing country, we decided to closely analyze the economic situation of Brazil, Germany, and South Africa two years before and after they hosted the World Cup. Moreover, there have been no previous research efforts that included factors beyond tourism and pure cost in an individual study or compared the host countries to non-host countries. Therefore, we have decided to include four factors within our research: GDP, unemployment, crime rates, and tourism rates in order to holistically assess the benefits or drawbacks that result after hosting the World Cup.

### Research Question

__We are interested in determining whether or not it is beneficial for a country to host the World Cup. To perform this analysis, we will analyze the GDP, unemployment, crime rates, and tourism in a host country two years before and after the World Cup was held there.__ Each factor will be examined in its original form and as a percent change from the previous year, to normalize variables across countries. We will compare the host country both to itself (pre-and post-World Cup), and to the rest of the world by using Spain, Morocco, and Mexico as our samples (one control country each from each respective continent: Europe, Africa, and the Americas). The three World Cups we will focus on are 1) the 2006 World Cup in Germany, 2) the 2010 World Cup in South Africa, and 3) the 2014 World Cup in Brazil. By comparing host countries' data to non-host countries' data, we aim to detect whether increases in prosperity are due to hosting the World Cup, or merely other confounding factors. As a sidenote, we did not examine the the 2018 World Cup in Russia because there is not yet data available for two years after the event.

### Hypothesis

We hypothesize that hosting a World Cup leads to __an increase in overall economic and social activity (increased GDP and tourism, decreased unemployment) but also leads to an increase in crime.__ We believe we will find that the benefits outweigh the costs and that hosting a World Cup has long-term positive impact on the host country.

Although we cannot directly measure long-term positive impact, we aim to provide insight for countries looking to host the World Cup by specifically utilizing Brazil (2014), South Africa (2010), and Germany (2006) as case studies for host countries. We will also be analyzing the same attributes for Mexico, Morocco, and Spain to represent non-host countries.  The specific attributes we will be examining are GDP, unemployment (as a % of the total labor force), crime rates (annual % change), and tourism (# of arrivals).

## Conducting T-tests on data

T-tests will be used to determine whether there is a *statistically significant* difference between tourism, unemployment, crime, and GDP rates before and after hosting the World Cup in Germany, South Africa, and Brazil. The test will be performed using mean percent change two years prior and post-World Cup in each country.

As mentioned above, our hypothesis is that hosting the World Cup leads to an increase in tourism, GDP, and crime, while also leading to a decrease in unemployment. For the purpose of the t-test, our associated null hypothesis, H<sub>0</sub>, is that hosting the World Cup had no effect on tourism/unemployment/crime/GDP.

The alternative hypothesis, H<sub>a</sub>, is that hosting the World Cup had an effect on tourism/unemployment/crime/GDP.

The p-value will be the primary factor analyzed from the test, in which a value under 0.05 may be used to reject the null hypothesis, while a value over 0.05 may be used to accept the null hypothesis. 

To put these results in perspective, however, the p-values from the three World Cup countries will be compared to those from the three control countries (Mexico, Morocco, and Spain), which represent the rest of the world. For the two years before and after each World Cup, the mean percent changes from the control countries will be used to determine the p-value for the rest of the world in the same time frame. For instance, if there is a significant difference in tourism in both the World Cup country and the rest of the world, it cannot be concluded that the jump in host country tourism is solely related to the world cup.
